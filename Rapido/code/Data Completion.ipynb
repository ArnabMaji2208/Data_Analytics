```python
import pandas as pd
```

```python
featured_data = pd.read_csv("../data/rapido_feature_engineered.csv")
```

```python
featured_data.head()
```

- target variable is missing

```python
holiday_data = pd.read_csv("../data/Holiday.csv")
holiday_data.DAY = holiday_data.DAY.apply(str)
holiday_data.DATE = holiday_data.DATE.astype('datetime64[ns]')
holiday_data.HOLIDAY = holiday_data.HOLIDAY.apply(str)
```

### User Id 69431

```python
selected_user = 69431
user = featured_data[featured_data.number==selected_user].reset_index(drop=True)
```

```python
user = user.drop_duplicates('date').reset_index(drop=True)
user['booking_class'] = 1
user.date = pd.to_datetime(user.date)
```

```python
negative_test_case = pd.DataFrame(data=pd.date_range(user.date.min(),user.date.max()),columns=['date'])
```

```python
negative_test_case['is_weekend'] = negative_test_case.date.dt.dayofweek.isin([5,6])
negative_test_case['is_weekday'] = negative_test_case.date.dt.dayofweek.isin(range(1,5))
negative_test_case['is_start_of_week'] = negative_test_case.date.dt.dayofweek.isin([0,1])
negative_test_case['is_wednesday'] = negative_test_case.date.dt.day_name().isin(['Wednesday'])
negative_test_case['is_end_of_week'] = negative_test_case.date.dt.dayofweek.isin([3,4])
```

```python
negative_test_case['is_holidays'] = negative_test_case.date.dt.date.isin(holiday_data.DATE.dt.date)
```

```python
negative_test_case['number']=selected_user
negative_test_case['booking_class']=0
```

```python
negative_test_case = negative_test_case[~negative_test_case.date.isin(user.date)]
```

```python
total_data_set =  pd.merge(user,negative_test_case,how='outer').reset_index(drop=True)
#total_data_set = pd.concat([user,negative_test_case])
```

```python
total_data_set.reindex(columns=["date","number","is_weekend","is_weekday","is_start_of_week","is_wednesday","is_end_of_week","is_holidays","booking_class"])
```

```python
total_data_set.drop(['number','date'],axis=1,inplace=True)
```

```python
total_data_set.is_weekday = total_data_set.is_weekday.astype('int')
total_data_set.is_start_of_week = total_data_set.is_start_of_week.astype('int')
total_data_set.is_weekend = total_data_set.is_weekend.astype('int')
total_data_set.is_end_of_week = total_data_set.is_end_of_week.astype('int')
total_data_set.is_wednesday = total_data_set.is_wednesday.astype('int')
total_data_set.is_holidays = total_data_set.is_holidays.astype('int')
```
